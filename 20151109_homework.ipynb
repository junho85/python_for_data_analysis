```python
# reference: https://wikidocs.net/65

import pandas as pd
from pylab import *

%matplotlib inline

x = linspace(-2, 2, 10000)
f = lambda x: sqrt(cos(x)) * cos(80*x) + 0.5*sqrt(abs(x))

plt.plot(x, f(x))
```
