### 선형회귀분석

```python
comp = np.array(['삼성전자', '현대차', '한국전력', '삼성물산', 'SK하이닉스'])
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
# [ 시가총액, 2/4분기매출, 2/4분기영업이익, 자산


data = np.array([[1870701, 485375, 68979, 2316909],
                 [356847, 228216, 17509,  1536161],
                 [310389, 136718, 20880, 1700829],
                 [300658, 13115, 391,  87722 ],
                 [273364, 46386, 13755, 284637 ]]
         )

                 
```

```python
lr = LinearRegression()
X = data[:,1:]
print(X.shape)
X
```

```python
y = data[:,0]
print(y.shape)
print(y)
```

```python
model = lr.fit(X, y)
model
```

```python
model.coef_
```

```python
#학습한 모델로 다시 예측해본 시가총액
# [ 삼성전자, 현대차, 한국전력, 삼성물산, SK하이닉스 ] 순
model.predict(X)
```

```python
# 모델의 정확도 평가. 1 이 perfect
model.score(X,y)
```

```python
# 모델의 정확도 평가. 분산값. 클수록 정확도가 나쁜 모델
np.mean((model.predict(X) - y) ** 2)
```
